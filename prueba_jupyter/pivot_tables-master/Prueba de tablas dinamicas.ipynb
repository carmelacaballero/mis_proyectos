```r
df <- read.csv("datos_competencia.csv")
df
```

```r
library(dplyr)
library(tidyr)
```

```r
df2 <- df %>% select(Nombre, Asignatura_completa, Nota, Competencia.matemática.y.competencias.básicas.en.ciencia.y.tecnología ) %>% group_by (Nombre, Asignatura_completa, Competencia.matemática.y.competencias.básicas.en.ciencia.y.tecnología) %>% summarise(Media=mean(Nota)) %>% filter(Competencia.matemática.y.competencias.básicas.en.ciencia.y.tecnología == "Sí")
```

```r
df2 %>% select(Nombre, Media) %>% group_by (Nombre) %>% summarize(Media = mean(Media), Adquirida_competencia = if(mean(Media)>=6.5){"SI"}else{"NO"})
```
