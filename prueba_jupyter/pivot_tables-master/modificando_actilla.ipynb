```python
def actilla(fich='importado1', eval=1):
    fichero = fich + '.csv'
    pre_actilla = pd.read_csv(fichero, index_col=False)
    pre_actilla = pre_actilla.drop([col for col in pre_actilla if col.startswith('Unna')], axis=1)
    pre_actilla = pd.melt(pre_actilla, id_vars=["N","Apellidos,Nombre"], var_name="Asignatura", value_name="Nota")
    pre_actilla = pre_actilla[pre_actilla['Nota'].notna()]
    pre_actilla = pre_actilla.copy()
    pre_actilla['Eval'] = eval       
    return pre_actilla
    

def concatenar_actillas(inicio_nombre_fichero='importado', evaluaciones_a_incluir=1): 
    return pd.concat([actilla(inicio_nombre_fichero + str(n + 1), n + 1) for n in range(evaluaciones_a_incluir)])


evaluaciones_a_incluir = 3
inicio_nombre_fichero = 'importado'


actilla_final = concatenar_actillas(inicio_nombre_fichero,evaluaciones_a_incluir)
print(actilla_final)
actilla_final.to_csv('exportado.csv', index=False, float_format='%.0f')




```

```python
# actilla_final.groupby(['Asignatura'])['N'].count().plot()
#actilla_final.groupby(['Asignatura'])['Nota'].mean().sort_values().plot(kind='bar')
df = actilla_final.groupby(['Asignatura', 'Eval'])['Nota'].mean().sort_values().to_frame()
df
```
