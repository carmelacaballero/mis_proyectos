### Repetir 100 veces el experimento de tirar cuatro de dados (o tirar cuatro veces elmismo dado) y contar el numero de ocasiones en las que se obtiene un 6.

```r
X=c(1,2,3,4,5,6)
Rep=rep(0,100)
Exitos=0
 for(i in 1:100)
    {M=sample(X,4,replace=TRUE)
     
     for(j in 1:4)
         {if(M[j]==6) Rep[i]=Rep[i]+1}
    if(Rep[i]>=1) Exitos=Exitos+1}
Rep
Exitos
          
```
